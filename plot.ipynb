```python
import matplotlib.pyplot as plt

categories = ['CNN', 'RN50', 'ViT-B/32', 'ViT-B/16']
v1 = [54.48, 52.01, 83.11, 83.16]
v2 = [90.53, 58.07, 48.00, 51.76]

colors = ['skyblue', 'salmon', 'lightgreen', 'gold']

# 创建柱状图
plt.bar(categories, v1, color=colors)


# 添加标题和标签
plt.title('Performance of classification task on CIFAR10')
plt.xlabel('Model')
plt.ylabel('Accuracy % ')


# 显示图表
plt.savefig('CIFAR10.png')
plt.show()


plt.bar(categories, v2, color=colors)

plt.title('Performance of classification task on MNIST')
plt.xlabel('Model')
plt.ylabel('Accuracy % ')

plt.savefig('MNIST.png')
plt.show()
```
