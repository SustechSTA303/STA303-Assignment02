```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('./data.csv')
data
```

```python
RN50 = data.iloc[np.r_[0:6, 7:11], 1:3]
ViTB16 = data.iloc[np.r_[11:17, 18:22], 1:3]
ViTB32 = data.iloc[np.r_[22:28, 29:33], 1:3]
ViTL14 = data.iloc[np.r_[33:39, 40:44], 1:3]


fig, ax = plt.subplots(figsize=(6, 4))


ax.plot(RN50.iloc[:,0], RN50.iloc[:,1], 'o--', label='RN50')

ax.plot(ViTB16.iloc[:,0], ViTB16.iloc[:,1], 'o--', label='ViT-B/16')

ax.plot(ViTB32.iloc[:,0], ViTB32.iloc[:,1], 'o--', label='ViT-B/32')

ax.plot(ViTL14.iloc[:,0], ViTL14.iloc[:,1], 'o--', label='ViT-L/14')

ax.set_ylabel('Accuracy(%)')
ax.set_title('FashionMNIST')

ax.legend(fontsize='small')
  

ax.tick_params(axis='x', rotation=90)

plt.tight_layout()

plt.savefig('./imagedata/FashionMNIST.png', dpi=300, bbox_inches='tight')

plt.show()

```
