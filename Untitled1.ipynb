2.MNIST dataset image classification

```python
text_inputs = torch.cat([clip.tokenize(f"grayscale image of a handwritten digit from the MNIST dataset with label {c}") for c in labels]).to(device)
def process_image(index, dataset, model, device, text):
    image, class_id = dataset[index]
    image_rgb = transforms.Grayscale(num_output_channels=3)(image)
    image_input = preprocess(Image.fromarray((image_rgb[0] * 255).byte().numpy())).unsqueeze(0).to(device)
    
    with torch.no_grad():
        image_features = model.encode_image(image_input)
        text_features = model.encode_text(text)

    image_features /= image_features.norm(dim=-1, keepdim=True)
    text_features /= text_features.norm(dim=-1, keepdim=True)
    return image_features, text_features
    #similarity = (100.0 * image_features @ text_features.T).softmax(dim=-1)
    #similarity = (100.0 * image_features @ text_features.T)
    #print(similarity[0].topk(1))
    #a, b = similarity[0].topk(1)
    #print(a.item())
    #print(b)
    # 
    #print("Image Features:", image_features)
    #print("Text Features:", text_features)
    #print("Similarity (softmax):", similarity)
    #return similarity[0].topk(1)
```

```python
def plot_image(image, class_id):
    plt.tight_layout()
    plt.imshow(image[0], interpolation='none')
    plt.title(f"label: {class_id}")
    plt.show()
```

```python
model, preprocess = clip.load(name=VISUAL_BACKBONE, device=device, download_root='/shareddata/clip/')
model.to(device)
```

```python
# Load the model

for i in range(100, 120):
    image, class_id = test_set[i]
    image_features, text_features = process_image(i, test_set, model, device, text_inputs)
    similarity = (100.0 * image_features @ text_features.T).softmax(dim=-1)
    bestvalue, bestindice = similarity[0].topk(1)
    truth = test_set.classes[bestindice].split(' ')
    print("True picture:")
    plot_image(image, class_id)
    print(f"\nTop predictions: {truth[0]}\n")
```

```python
visual_backbones = ['RN101','RN50','RN50x4', 'RN50x16', 'RN50x64','ViT-B/32','ViT-L/14@336px','ViT-L/14', 'ViT-B/16'  ]
#visual_backbones = ['RN101']
accuracy_scores = []
for visual in visual_backbones:
    VISUAL_BACKBONE = visual
    model, preprocess = clip.load(name=VISUAL_BACKBONE, device=device, download_root='/shareddata/clip/')
    model.to(device)
    correct = 0
    for i in range(len(test_set)):
        image, class_id = test_set[i]
        image_features, text_features = process_image(i, test_set, model, device, text_inputs)
        similarity = (100.0 * image_features @ text_features.T).softmax(dim=-1)
        bestvalue, bestindice = similarity[0].topk(1)
        truth = test_set.classes[bestindice].split(' ')
    #print(truth[0])
    #print(class_id)
        if int(class_id) == int(truth[0]):
            correct = correct + 1
    val_acc = correct / 60000 * 100
    accuracy_scores.append(val_acc)
    print(f"the zero-shot performance on {dataset_name} is {val_acc:.2f}%, visual encoder is {VISUAL_BACKBONE}.")
```

```python
plt.plot(visual_backbones, accuracy_scores, marker='o', linestyle='-')

# 
plt.title('CLIP Model Accuracy with Different Visual Backbones')
plt.xlabel('Visual Backbone')
plt.ylabel('Accuracy')

# 
plt.show()
```
