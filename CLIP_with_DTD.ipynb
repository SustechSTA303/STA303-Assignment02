```python
import os
import time
import os.path as osp

import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader

from torchvision import datasets
from torchvision import transforms
import torchvision

from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
from PIL import Image
from clip import clip
```

```python
# random seed
SEED = 1 
NUM_CLASS = 10

# Training
BATCH_SIZE = 128
NUM_EPOCHS = 30
EVAL_INTERVAL=1
SAVE_DIR = './log'

# Optimizer
LEARNING_RATE = 1e-1
MOMENTUM = 0.9
STEP=5
GAMMA=0.5
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
transform_dtd = transforms.Compose([
    transforms.Resize(size=224),
    transforms.CenterCrop(size=(224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
    # 如果需要，可以添加归一化参数
])

# DTD 数据集的文件夹路径
dtd_path = '/shareddata/dtd/dtd/images'

# 加载 DTD 数据集
dtd_dataset = datasets.ImageFolder(root=dtd_path, transform=transform_dtd)
dtd_dataloader = torch.utils.data.DataLoader(dtd_dataset, batch_size=BATCH_SIZE, shuffle=True)
```

```python
# 提取所有子文件夹名称作为类名
class_names = [d for d in os.listdir(dtd_path) if os.path.isdir(os.path.join(dtd_path, d))]
print("Class names in DTD dataset:", class_names)
dataset_name = 'dtd'
```

```python
text_inputs = torch.cat([clip.tokenize(f"A photo of a {class_name}").to(device) for class_name in class_names], dim=0)
```

```python
def model_inference(model, image):
    
    image_features = model.encode_image(image)
    text_features = model.encode_text(text_inputs)

    image_features /= image_features.norm(dim=-1, keepdim=True)
    text_features /= text_features.norm(dim=-1, keepdim=True)

    logit_scale = model.logit_scale.exp()

    logits = logit_scale * image_features @ text_features.t()

    return logits
```

```python
torch.cuda.empty_cache()
```

```python
VISUAL_BACKBONE = 'RN50'
# Load the model
model, preprocess = clip.load(name=VISUAL_BACKBONE, device=device, download_root='/shareddata/clip/')
model.to(device)
```

```python
with torch.no_grad():
    model.eval()

    val_loss = 0.0
    val_corrects = 0

    for batch_idx, (image, target) in enumerate(dtd_dataloader):

        image = image.to(device)
        target = target.to(device)

        # test model
        logits = model_inference(model, image)
        _, preds = torch.max(logits, 1)
        
        val_corrects += torch.sum(preds == target.data)

    val_acc = val_corrects.double() / len(dtd_dataset)

    print(f"the zero-shot performance on {dataset_name} is {val_acc*100:.2f}%, visual encoder is {VISUAL_BACKBONE}.")
acc1 = val_acc
```
