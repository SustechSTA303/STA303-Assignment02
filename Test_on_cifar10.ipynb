```python
# # random seed
# SEED = 1 
# NUM_CLASS = 10

# Training
BATCH_SIZE = 128
# NUM_EPOCHS = 30
# EVAL_INTERVAL=1
# SAVE_DIR = './log'

# # Optimizer
# LEARNING_RATE = 1e-1
# MOMENTUM = 0.9
# STEP=5
# GAMMA=0.5

# CLIP
VISUAL_BACKBONE = 'RN50' # RN50, ViT-B/32, ViT-B/16
```

```python
transform = transforms.Compose([
    transforms.Resize(size=224),
    transforms.CenterCrop(size=(224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
```

```python
#device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device="cpu"
# Load the model
model, preprocess = clip.load(name=VISUAL_BACKBONE, device=device, download_root='/shareddata/clip/')
model.to(device)
```

```python
class_descriptions = testset.classes # 每个 CIFAR100 类别的描述
class_descriptions
# class_descriptions = ['a photo of ' + animal for animal in class_descriptions]
class_descriptions
```

```python
def classify_images(model, data_loader, class_descriptions, device):
    model.eval()
    correct = 0
    total = 0

    # 将类别描述转换为文本特征
    text_tokens = clip.tokenize(class_descriptions).to(device)
    with torch.no_grad():
        text_features = model.encode_text(text_tokens)

    with torch.no_grad():
        for images, labels in data_loader:
            images = images.to(device)
            labels = labels.to(device)
          #  print(labels)
            # 提取图像特征
            image_features = model.encode_image(images)
            
            # 计算图像特征与每个类别文本特征之间的相似度
            similarity = (image_features @ text_features.T).softmax(dim=-1)

            # 获取最高相似度的类别作为预测
            predictions = similarity.argmax(dim=-1)

            correct += (predictions == labels).sum().item()
            total += labels.size(0)

    return correct / total
```

```python
# 4. 性能评估
accuracy = classify_images(model, testloader, class_descriptions, device)
print(f'Accuracy on CIFAR10 test images: {accuracy * 100:.2f}%')
```
