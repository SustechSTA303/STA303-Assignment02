```python
prompt = 'a photo of a ' # you can try different prompt

def prompt_encode(prompt):
    """
    Args:
        prompt (str): the text prefix before the class

    Returns:
        text_inputs(torch.Tensor)

    """
    ##################### Write your answer here ##################
    # 生成所有候选文本
    text = [f"{prompt} {class_name}" for class_name in class_names]
    
    # 使用CLIP文本编码器对文本进行编码
    text_inputs = clip.tokenize(text).to(device)

    ###############################################################
    
    return text_inputs
```

### Task 2: Zero-shot inference
Please denfine a function named model_inference. The function is essential for training and evaluating machine learning models using batched data from dataloaders.

**To do**:

1. Encode the image.
2. Encode the text.
3. Calculate the logits.
---

```python
def model_inference(model, image, text_inputs):
    """
    Args:
        model (torch.nn.Module): The CLIP model.
        image (torch.Tensor): A batch of images.
        text_inputs (torch.Tensor): Encoded text inputs.

    Returns:
        logits_per_image (torch.Tensor), logits_per_text (torch.Tensor)
    """
    # 使用CLIP模型的图像编码器对图像进行编码
    image_features = model.encode_image(image)
    
    # 使用CLIP模型的文本编码器对文本进行编码
    text_features = model.encode_text(text_inputs)
    
    # 直接使用模型对图像和文本进行编码，并计算logits
    logits_per_image, _ = model(image, text_inputs)
    
    return logits_per_image
```

```python
# 初始化一个空列表来保存每个epoch的准确度
accuracy_list = []

# 对每个epoch进行循环
for epoch in range(NUM_EPOCHS):
    correct = 0
    total = 0

    # 对测试数据集中的每个batch进行循环
    for images, labels in test_dataloader:
        # 将图像和标签移动到设备上
        images = images.to(device)
        labels = labels.to(device)

        # 生成提示
        text_inputs = prompt_encode(prompt)

        # 进行模型推理
        logits = model_inference(model, images, text_inputs)

        # 计算预测的类别
        predictions = torch.argmax(logits, dim=-1)

        # 更新正确预测的数量和总数量
        correct += (predictions == labels).sum().item()
        total += labels.size(0)

    # 计算精度
    val_acc = correct / total
    accuracy_list.append(val_acc)

    print(f"Epoch {epoch+1}: the zero-shot performance on {dataset_name} is {val_acc*100:.2f}%, visual encoder is {VISUAL_BACKBONE}.")

# 绘制准确度随epoch变化的图像
plt.figure(figsize=(10, 5))
plt.plot(range(1, NUM_EPOCHS+1), accuracy_list, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Epochs')
plt.grid(True)
plt.show()
```

```python
criterion = nn.CrossEntropyLoss()

# 定义优化器
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# 初始化一个空列表来保存每个批次的损失
accuracy_list = []
loss_list = []

# 对每个epoch进行循环
for epoch in range(NUM_EPOCHS):
    correct = 0
    total = 0

    # 对测试数据集中的每个batch进行循环
    for images, labels in test_dataloader:
        # 将图像和标签移动到设备上
        images = images.to(device)
        labels = labels.to(device)

        # 生成提示
        text_inputs = prompt_encode(prompt)

        # 进行模型推理
        logits = model_inference(model, images, text_inputs)

        # 计算损失
        loss = criterion(logits, labels)

        # 添加损失到损失列表
        loss_list.append(loss.item())

        # 反向传播和优化
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # 计算预测的类别
        predictions = torch.argmax(logits, dim=-1)

        # 更新正确预测的数量和总数量
        correct += (predictions == labels).sum().item()
        total += labels.size(0)

    # 计算精度
    val_acc = correct / total
    accuracy_list.append(val_acc)

    print(f"Epoch {epoch+1}: the zero-shot performance on {dataset_name} is {val_acc*100:.2f}%, visual encoder is {VISUAL_BACKBONE}.")

# 绘制损失随批次变化的图像
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(loss_list)+1), loss_list, marker='o')
plt.xlabel('Batches')
plt.ylabel('Loss')
plt.title('Loss vs. Batches')
plt.grid(True)
plt.show()

# 绘制准确度随epoch变化的图像
plt.figure(figsize=(10, 5))
plt.plot(range(1, NUM_EPOCHS+1), accuracy_list, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Epochs')
plt.grid(True)
plt.show()
```
