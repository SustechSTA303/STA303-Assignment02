```python
#展示MINST的一些示例
transform = transforms.Compose([transforms.ToTensor(),
                                transforms.Grayscale(num_output_channels=3),
                               ])
batch_size = 10
mnist_train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
mnist_test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
mnist_train_loader = torch.utils.data.DataLoader(mnist_train_dataset, batch_size=batch_size, shuffle=True)
images, labels = next(iter(mnist_train_loader))
fig, axes = plt.subplots(1, batch_size, figsize=(15, 3))
for i in range(batch_size):
    axes[i].imshow(images[i].squeeze(), cmap='gray')
    axes[i].set_title(f"Label: {labels[i]}")
    axes[i].axis('off')
plt.show()
print("MINST Training Set Size:",len(mnist_train_dataset))
print("MINST Testing Set Size:", len(mnist_test_dataset))
print("CMINST Diversity:", len(mnist_train_dataset.classes))
```

```python
transform_cifar100_test = transforms.Compose([
    transforms.ToTensor(),
])

# 加载 CIFAR-100 训练集（如果没有下载，会自动下载）
train_dataset2 = datasets.CIFAR100(root='./data', train=True, download=True,transform=transform_cifar100_test )
test_dataset2 = datasets.CIFAR100(root='./data', train=False, download=True,transform=transform_cifar100_test)

cifar100_train_dataset = train_dataset2
batch_size = 5
cifar100_train_loader = torch.utils.data.DataLoader(cifar100_train_dataset, batch_size=batch_size, shuffle=True)
images, labels = next(iter(cifar100_train_loader))
class_names = cifar100_train_dataset.classes
fig, axes = plt.subplots(1, batch_size, figsize=(15, 3))
for i in range(batch_size):
    class_label = class_names[labels[i]] if class_names else str(labels[i])
    axes[i].imshow(images[i].permute(1, 2, 0))  
    axes[i].set_title(f"Label: {class_label}")
    axes[i].axis('off')
plt.show()

print("CIFAR-100 Training Set Size:",len(train_dataset2))
print("CIFAR-100 Testing Set Size:", len(test_dataset2))
print("CIFAR-100 Diversity:", len(train_dataset2.classes))
```

```python
transform_cifar10_test = transforms.Compose([
    transforms.ToTensor(),
])

# 加载 CIFAR-10 训练集（如果没有下载，会自动下载）
train_dataset1 = datasets.CIFAR10(root='./data', train=True, download=True,transform=transform_cifar10_test )
test_dataset1 = datasets.CIFAR10(root='./data', train=False, download=True,transform=transform_cifar10_test)

cifar10_train_dataset = train_dataset1
batch_size = 5
cifar10_train_loader = torch.utils.data.DataLoader(cifar10_train_dataset, batch_size=batch_size, shuffle=True)
images, labels = next(iter(cifar10_train_loader))
class_names = cifar10_train_dataset.classes
fig, axes = plt.subplots(1, batch_size, figsize=(15, 3))
for i in range(batch_size):
    class_label = class_names[labels[i]] if class_names else str(labels[i])
    axes[i].imshow(images[i].permute(1, 2, 0))  
    axes[i].set_title(f"Label: {class_label}")
    axes[i].axis('off')
plt.show()

print("CIFAR-100 Training Set Size:",len(train_dataset1))
print("CIFAR-100 Testing Set Size:", len(test_dataset1))
print("CIFAR-100 Diversity:", len(train_dataset1.classes))
```

```python
#展示Fashion-MINST的一些示例
transform = transforms.Compose([transforms.ToTensor(),
                                transforms.Grayscale(num_output_channels=3),
                               ])
batch_size = 10
mnist_train_dataset = datasets.FashionMNIST(root='./data', train=True, download=True, transform=transform)
mnist_test_dataset = datasets.FashionMNIST(root='./data', train=False, download=True, transform=transform)
mnist_train_loader = torch.utils.data.DataLoader(mnist_train_dataset, batch_size=batch_size, shuffle=True)
images, labels = next(iter(mnist_train_loader))
fig, axes = plt.subplots(1, batch_size, figsize=(15, 3))
for i in range(batch_size):
    axes[i].imshow(images[i].squeeze(), cmap='gray')
    axes[i].set_title(f"Label: {labels[i]}")
    axes[i].axis('off')
plt.show()
print("FashionMINST Training Set Size:",len(mnist_train_dataset))
print("FashionMINST  Testing Set Size:", len(mnist_test_dataset))
print("FashionMINST  Diversity:", len(mnist_train_dataset.classes))
```
